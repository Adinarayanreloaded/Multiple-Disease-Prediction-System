```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
ds = pd.read_csv('breast cancer.csv')
ds.head()
```

```python
ds.tail()
```

```python
x = ds.drop(columns = ['id', 'diagnosis' ,'Unnamed: 32'], axis = 1)
y = ds['diagnosis']
```

```python
print(x)
```

```python
print(y)
```

```python
ds.shape
```

```python
ds.describe()
```

```python
y.value_counts()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
ds2 = scaler.transform(x)
```

```python
x = ds2
```

```python
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
print(x_train)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
#training the support vector machine classifier
classifier.fit(x_train, y_train)
```

```python
x_train_predict = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predict, y_train)
```

```python
print('accuracy of training dataset :: ', training_data_accuracy)
```

```python
x_test_predict = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predict, y_test)
```

```python
print('accuracy of testing dataset :: ', test_data_accuracy)
```
